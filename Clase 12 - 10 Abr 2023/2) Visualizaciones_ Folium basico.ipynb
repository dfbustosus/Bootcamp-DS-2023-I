```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/JJTorresDS/ds-data-sources/main/properati_venta_201501.csv')
df
```

```python
# Separar la columna lat/lon
df[['lat', 'lon']]=df['lat_lon'].str.split(',', expand=True)
df.head()
```

```python
df['lat']=df.lat.astype(float)
df['lon']=df.lon.astype(float)
```

```python
df.info()
```

```python
import folium  #needed for interactive map
from folium.plugins import HeatMap
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df= df.dropna()
df.isnull().sum()
```

```python
df.shape
```

```python
# Precio
max_amount = float(df['price_usd_per_m2'].max())

folium_hmap = folium.Map(location=[-40, -70],
                        zoom_start=5,
                        tiles="OpenStreetMap")

hm_wide = HeatMap( list(zip(df['lat'], df['lon'], df['price_usd_per_m2'])),
                   min_opacity=0.8,
                   #max_val=max_amount,
                   radius=10, blur=6, 
                   max_zoom=15, 
                 )

folium_hmap.add_child(hm_wide)
```

```python
# Area
max_amount = float(df['surface_total_in_m2'].max())

folium_hmap = folium.Map(location=[-40, -70],
                        zoom_start=5,
                        tiles="OpenStreetMap")

hm_wide = HeatMap( list(zip(df['lat'], df['lon'], df['surface_total_in_m2'])),
                   min_opacity=0.8,
                   #max_val=max_amount,
                   radius=10, blur=6, 
                   max_zoom=15, 
                 )

folium_hmap.add_child(hm_wide)
```
